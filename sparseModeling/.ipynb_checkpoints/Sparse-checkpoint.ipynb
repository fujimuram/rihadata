```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%load_ext autoreload
%autoreload 1

%config InlineBackend.figure_formats = {'png', 'retina'}
```

```python
%aimport matrix
from PIL import Image
fname = './input/hosi.png'
size = 100
w, h = size,size
img = Image.open(fname)
img_resize_lanczos = img.resize((h,w), Image.LANCZOS)
imG = np.array(img_resize_lanczos)
#imG = np.array(img_resize_lanczos.convert('L'))
```

```python
imG
```

```python
vmax = np.ceil(np.max(imG))

plt.figure(figsize=(8,6))
#ax = sns.heatmap(imG, vmin=-1, vmax=vmax)

plt.savefig('output/original.png')
```

```python
drop_ratio = 0.5
imG_dropped = matrix.drop_values(imG, ratio=drop_ratio)

plt.figure(figsize=(9,6))
ax = sns.heatmap(imG_dropped, vmin=-1, vmax=vmax)
ax.set_title(f'{drop_ratio:.0%} dropped')

plt.savefig('output/dropped.png')
```

```python
%%time

from sklearn.metrics import mean_squared_error

model = matrix.LowrankReconstruction()#低ランク行列再構成
imG_reconstructed = model.transform(imG_dropped)
#imG_reconstructed = model.transform(imG)
mse = mean_squared_error(imG, imG_reconstructed)

plt.figure(figsize=(8,6))
ax = sns.heatmap(imG_reconstructed, vmin=-1, vmax=vmax)
ax.set_title(f'reconstructed (MSE {mse:.6f})')

plt.savefig('output/reconstructed.png')
print(mse)
```

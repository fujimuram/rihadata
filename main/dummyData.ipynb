```python
import numpy as np
from scipy.stats import norm
import math
import statistics
import random
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score

def printmap(lst):
    for l in lst:
        A=[str(v) for v in l]
        print(''.join(A))
def maxone(Y):
    return Y/maxY
```

```python
#0から100まで、0.01間隔で入ったリストXを作る！
X = np.arange(0,w,1)
#確率密度関数にX,平均、標準偏差を代入
Y = norm.pdf(X,mean,sigma)
maxY= max(Y)
# map関数を使用してdata1の要素を全て倍にしてdata2に格納する場合
Y = list(map(maxone, Y))# [2, 6, 12, 100, 10]
plt.plot(X,Y,color='r')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import time

start_time = time.time()
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
# 小数化
x_train = x_train / 2
x_test = x_test / 2
# データ数
m_train, m_test = x_train.shape[0], x_test.shape[0]
# ベクトル化
x_train, x_test = x_train.reshape(m_train, -1), x_test.reshape(m_test, -1)
print(x_train.shape)
"""# ノルムで標準化
x_train = x_train / np.linalg.norm(x_train, ord=2, axis=1, keepdims=True)
x_test = x_test / np.linalg.norm(x_test, ord=2, axis=1, keepdims=True)"""

# RF
clf = RandomForestRegressor()
clf.fit(x_train, y_train)
print("Elapsed[s] : ", time.time() - start_time)
print("Train :", clf.score(x_train, y_train))
print("Test :", clf.score(x_test, y_test))
```
